## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df_cds['line_subtotal']=df_cds['unit_price']*df_cds['qty']
df_cds[['unit_price','qty','line_subtotal']].head(5)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# created a computed column total_weight 
df_cds['total_weight']=df_cds['unit_weight'] * df_cds['qty'] 

df_cds['shipping_price'] = df_cds.apply(lambda rw: rw['total_weight'] * 7 if rw['total_weight'] > 50 else rw['total_weight'] * 10,axis=1)

df_cds[['unit_price','unit_weight','qty','total_weight','shipping_price']].head(5)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df_cds['line_price']=round((df_cds['line_subtotal'] + df_cds['shipping_price'])*(1.0925),2)

df_cds[['line_subtotal','shipping_price','line_price']].head(5)
# ** other columns exist but are not displayed 
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df_cds['line_cost']=round((df_cds['unit_cost']*df_cds['qty']) + df_cds['shipping_price'],2)

df_cds[['shipping_price','line_price','unit_cost','qty','line_cost']].head(5)
# ** other columns exist but are not displayed 
```

```python
# Create a column for the profit of
# each line using line cost and line price

df_cds['line_profit']=df_cds['line_price']-df_cds['line_cost']

df_cds[['client_id','qty','line_price','line_cost','line_profit']].head(5)

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

df_cds[df_cds['client_id'].isin(top5bycnt)].groupby(df_cds['client_id'])['line_price'].sum().sort_values().map('${:,.2f}'.format)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

df_summary = df_cds[df_cds['client_id'].isin(top5bycnt)].groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'line_price': 'sum',
    'line_cost': 'sum',
    'line_profit' :'sum'
}).sort_values('line_profit',ascending=False)

df_summary
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
df_summary_Formatted=df_summary.apply(lambda col: col.apply(lambda cel: '${:,.2f}M'.format(cel / 10**6) if col.name !='qty' else cel ))
df_summary_Formatted.columns=['Units','Shipping','Total Revenue','Total Cost','Total Profit']
df_summary_Formatted
#'Client ID','Units','Shipping','Total Revenue','Total Cost','Total Profit'
```

```python
# Sort the updated data by 'Total Profit' form highest to lowest

# the data set is already sorted 
# but if it was not and  we wanted to sort again 
# we would have toconvert total profit which is a string column to a digit. 
df_summary_Formatted['ntp']= pd.to_numeric(df_summary_Formatted['Total Profit'].str.replace('$','').str.replace('M',''))

df_summary_Formatted.sort_values('ntp',ascending=False).filter(['Client ID','Units','Shipping','Total Revenue','Total Cost','Total Profit'])

```
